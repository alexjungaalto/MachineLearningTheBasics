```python
print(__doc__)

# Author: Phil Roth <mr.phil.roth@gmail.com>
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

n_samples = 20
random_state = 20
X, y = make_blobs(n_samples=n_samples, random_state=random_state)
x1=X[:, 0]
x2=X[:,1]
 # class sklearn.cluster.KMeans(n_clusters=8, *, init='k-means++', n_init=10, 
#                              max_iter=300, tol=0.0001, precompute_distances='deprecated', 
  #                            verbose=0, random_state=None, copy_x=True, n_jobs='deprecated', 
   # algorithm='auto')[source]Â¶
# Incorrect number of clusters
mc = np.random.rand(2,2)
kmeansobj = KMeans(n_clusters=2,init=mc,max_iter=1)
y_pred = kmeansobj.fit_predict(X)

plt.subplot(221)
#cluster1 = y_pred==1
#x1=X[:, 0]
#x1_cluster1=x1[clusteridx]
#cluster1 = X[clusteridx,:]
#print(cluster1)
plt.scatter(x1[y_pred==0], x2[y_pred==0], marker='x',s=100)
plt.scatter(x1[y_pred==1], x2[y_pred==1], marker='o',s=100)
mc = kmeansobj.cluster_centers_
plt.scatter(mc[0,0],mc[0,1],marker='x',s=400)
plt.scatter(mc[1,0],mc[1,1],marker='o',s=400)
plt.title("first iteration")

# Anisotropicly distributed data
#transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
#X_aniso = np.dot(X, transformation)
#y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)
print(mc.shape)
kmeansobj = KMeans(n_clusters=2, init=mc,max_iter=1)
y_pred = kmeansobj.fit_predict(X)

plt.subplot(222)
plt.scatter(x1[y_pred==0], x2[y_pred==0], marker='x',s=100)
plt.scatter(x1[y_pred==1], x2[y_pred==1], marker='o',s=100)
mc = kmeansobj.cluster_centers_
plt.scatter(mc[0,0],mc[0,1],marker='x',s=400)
plt.scatter(mc[1,0],mc[1,1],marker='o',s=400)
plt.title("second iteration")


kmeansobj = KMeans(n_clusters=2, init=mc,max_iter=1)
y_pred = kmeansobj.fit_predict(X)

plt.subplot(223)
plt.scatter(x1[y_pred==0], x2[y_pred==0], marker='x',s=100)
plt.scatter(x1[y_pred==1], x2[y_pred==1], marker='o',s=100)
mc = kmeansobj.cluster_centers_
plt.scatter(mc[0,0],mc[0,1],marker='x',s=400)
plt.scatter(mc[1,0],mc[1,1],marker='o',s=400)
plt.title("third iteration")


kmeansobj = KMeans(n_clusters=2, init=mc,max_iter=1)
y_pred = kmeansobj.fit_predict(X)

plt.subplot(224)
plt.scatter(x1[y_pred==0], x2[y_pred==0], marker='x',s=100)
plt.scatter(x1[y_pred==1], x2[y_pred==1], marker='o',s=100)
mc = kmeansobj.cluster_centers_
plt.scatter(mc[0,0],mc[0,1],marker='x',s=400)
plt.scatter(mc[1,0],mc[1,1],marker='o',s=400)
plt.title("fourth iteration")
plt.savefig('EvolKmeans.eps', format='eps')
plt.show()
```
